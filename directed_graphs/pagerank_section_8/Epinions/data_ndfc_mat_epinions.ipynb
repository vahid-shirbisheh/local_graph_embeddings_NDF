```python
from dihelpers import *
import pandas as pd
```

# Importing Graph Data from CSV File

```python
d = pd.read_csv("epinions.csv", delimiter='\t')
edge_list = [(x,y) for [x,y] in d.values]
mydg0 = MyDiGraph()
mydg0.from_edge_list(edge_list) # Constructing the graph using the list of its edges
nodes = mydg0.get_vertex_list()
nodes.sort()
nodes[0:10]
```

```python
mydg = MyDiGraph()
for node in nodes: # Creating nodes in the ordinary order
    mydg.add_vertex(node)
mydg.from_edge_list(edge_list) # Constructing the graph using the list of its edges
```

```python
mydg.in_degrees_list()
```

# The NDFC matrix representation

```python
sp_1 = mydg.starting_points(ratio=1.15, max_length=70, starting_length=1, inward=True, last_point=980)
print(len(sp_1))
sp_1
```

```python
sp_1 = sp_1 + [1100, 1200, 1450, 2800]
sp_1
```

```python
%%time
# producing the order 3 NDFC matrix rep. of nodes of the garph w.r.t. the starting points sp_1:
mats_1 = NDFC_matrix_rep(mydg, starting_points=sp_1, inward=True, radius=3, decimals=6)
mats_1[0]
```

```python
mats_1.shape
```

```python
np.save("epinions_NDFC_matrix_r1_15_sta1_max70_rad3_inward.npy", mats_1, allow_pickle=False, fix_imports=True)
```
