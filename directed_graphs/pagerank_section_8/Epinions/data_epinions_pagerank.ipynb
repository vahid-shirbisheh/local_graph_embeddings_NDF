```python
from dihelpers import *
import pandas as pd
```

# Importing Graph Data from CSV File

```python
d = pd.read_csv("epinions.csv", delimiter='\t')
d.head(3)
```

```python
edge_list = [(x,y) for [x,y] in d.values]
mydg0 = MyDiGraph()
mydg0.from_edge_list(edge_list) # Constructing the graph using the list of its edges
nodes = mydg0.get_vertex_list()
nodes.sort()
nodes[0:10]
```

```python
mydg = MyDiGraph()
for node in nodes: # Creating nodes in the ordinary order
    mydg.add_vertex(node)
mydg.from_edge_list(edge_list) # Constructing the graph using the list of its edges
nodes1 = mydg.get_vertex_list()
nodes1[0:10]
```

```python
%%time
nxdg = nx.DiGraph()
nxdg.add_nodes_from(mydg.get_vertex_list())
nxdg.add_edges_from(edge_list)
```

```python
nodes2 = list(nx.nodes(nxdg))
nodes2[0:10]
```

# PageRank

Producing the PageRanks of the nodes of the graph using the library NetworkX

```python
%%time
# The array of pagerank of vertices of the graph
np_target_page = nx.algorithms.link_analysis.pagerank_alg.pagerank(nxdg)
np_target_page
```

```python
%%time
# The array of pagerank of vertices of the graph
np_target_page = np.array(list(nx.algorithms.link_analysis.pagerank_alg.pagerank(nxdg).values())).reshape(-1,1)
np_target_page
```

```python
# Saving the pagerank as a numpy array into a file
np.save("epinions_pageranks.npy", np_target_page, allow_pickle=False, fix_imports=True)
```
