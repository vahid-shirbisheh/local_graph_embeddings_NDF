```python
mydg = MyDiGraph()
nodes = ["A", "H", "B", "Y", "I", "C", "L", "E", "D", "J", "M", "K", "N", "Q", "R", "P"]
edges = [("B", "E"), ("K", "N"), ("F", "C"), ("B", "A"), ("A", "C"), ("C", "F"), ("F", "E"), ("L", "Y"), ("Y", "A"),
         ("A", "H"), ("H", "M"), ("M", "N"), ("K", "I"), ("I", "H"), ("H", "Y"), ("Y", "I"), ("A", "J"), ("B", "D"),
         ("D", "E"), ("Y", "L"), ("L", "K"), ("Q", "D"), ("D", "R"), ("P", "K"), ("K", "P")]
for node in nodes:
    mydg.add_vertex(node)
mydg.from_edge_list(edges)
```

```python
dg(mydg, "kamada")
```

```python
print("The list of in-degrees:  ", mydg.in_degrees_list())
print("The list of out-degrees: ", mydg.out_degrees_list())
```

# Vanilla NDF vectors (inward/outward) 

```python
sp = [0, 1, 2, 3]
for node in mydg.adj_list:
    print(f"{node:{3}}  in-degree: {mydg.in_deg(node):{5}}              out-degree: {mydg.out_deg(node):{5}}")
    print(f"Vanilla in-NDF: {mydg.ndf(node, starting_points=sp, inward=True)}  Vanilla out-NDF: {mydg.ndf(node, starting_points=sp, inward=False)}")
    print("______________________________________________________")
```
