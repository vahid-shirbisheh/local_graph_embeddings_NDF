# Visualization of two nodes A and Q

Figures 2 and Example 5.3 in the article are based on this notebook

```python
from mygraph import MyGraph
from helpers import *
```

```python
# Building the graph 
myg = MyGraph()
edge_list = [("A", "B"), ("A", "C"), ("A", "J"), ("A", "Y"), ("A", "H"), ("Y", "I"), ("Y", "L"), ("Y", "H"),  
             ("I", "K"), ("B", "E"), ("B", "D"), ("D", "R"), ("D", "Q"), ("C", "F"), ("F", "E"), ("E", "D"), 
             ("L", "K"), ("K", "P"), ("H", "I"), ("H", "M"), ("M", "N"), ("N", "K") ]
myg.from_edge_list(edge_list)
```

```python
# drawing the graph
dg(myg)
```

```python
sp = myg.starting_points(ratio=0.2)
VNDFC_mats = VNDFC_matrix_rep(myg, starting_points=sp, radius=7, decimals=3)
```

```python
nodes = list(myg.adj_list)
```

```python
for i, node in enumerate(nodes):
    print(node)
    print(VNDFC_mats[i])
    figure(figsize=(3, 3), dpi=80)
    plt.imshow(VNDFC_mats[i])
    plt.show()
```

```python
for node in ["A", "C", "J", "D", "Q", "M", "N", "B", "E"]:
    print(node)
    i = nodes.index(node)
    #print(VNDFC_mats[i])
    figure(figsize=(3, 3), dpi=60)
    plt.imshow(VNDFC_mats[i], cmap='gray')
    plt.show()
```
